### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"There are {count} establishments with a hygiene score of 20")

# Display the first document in the results using pprint
first_result = establishments.find_one(query)
pprint(first_result)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
from pprint import pprint

# Find establishments in London, which is divided into multiple Local Authorities, with a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex':'London'}, 'RatingValue': {'$gte': 4}}  # Remove quotes around 4

# Count the documents that match the query
count = establishments.count_documents(query)

# Print the count of documents
print(f"Number of documents in the result: {count}")

# Find and display the first document in the results using pprint
if count > 0:
    result = establishments.find(query).next()
    pprint(result)
else:
    print("No matching documents found.")
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
from pprint import pprint

# Search for "Penang Flavours"
query = {'BusinessName': 'Penang Flavours'}
results = establishments.find(query)

# Print the results for "Penang Flavours"
for result in results:
    pprint(result)

# Define the search parameters
degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

# Define the query to find establishments within the specified radius with a RatingValue of 5
query = {
    'geocode.latitude': {'$lte': latitude + degree_search, '$gte': latitude - degree_search},
    'geocode.longitude': {'$lte': longitude + degree_search, '$gte': longitude - degree_search},
    'RatingValue': 5  # Use an integer instead of a string
}

# Define the sort order by hygiene score
sort = [('scores.Hygiene', 1)]

# Limit the number of results to 5
limit = 5

# Find establishments that match the criteria and sort by hygiene score
results = list(establishments.find(query).sort(sort).limit(limit))

# Print the results
pprint(results)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
import pandas as pd
from pprint import pprint

# Define the pipeline stages
pipeline = [
    {
        '$match': {
            'scores.Hygiene': 0
        }
    },
    {
        '$group': {
            '_id': '$LocalAuthorityName',
            'count': {'$sum': 1}
        }
    },
    {
        '$sort': {
            'count': -1
        }
    }
]

# Execute the pipeline using the aggregate method, cast the results as a list, and store them in a variable
results = list(establishments.aggregate(pipeline))

# Convert the results to a Pandas DataFrame
result_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_df))

# Display the first 10 rows of the DataFrame
pprint(result_df.head(10))
```
